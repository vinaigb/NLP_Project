## Model: "sshleifer/distilbart-cnn-12-6"

sshleifer/distilbart-cnn-12-6 is a distilled version of the BART model trained specifically for text summarization. It utilizes a convolutional neural network (CNN) architecture with 12 encoder and 6 decoder layers to encode the input text and generate concise summaries.

```python
from transformers import pipeline
import os

# Define the file path of the input text file
file_path = "/content/drive/My Drive/dataset/sampleData/combined_output.txt"

# Initialize the summarization pipeline with model name and revision
summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6",
    revision="a4f8f3e"
)

# Create an empty list to store the summaries
summaries = []

# Read the content of the input text file
with open(file_path, "r") as file:
    article = file.read()
    # Truncate the article to fit within the maximum sequence length
    max_length = summarizer.tokenizer.model_max_length - summarizer.tokenizer.num_special_tokens_to_add(pair=True)
    article = article[:max_length]
    # Generate the summary for the input file
    summary = summarizer(article, max_length=150, min_length=100, do_sample=False)[0]['summary_text']
    summaries.append(summary)

# Define the output file path
output_file_path = "/content/drive/My Drive/dataset/sampleData/summary2.txt"

# Save the summary to the output text file
with open(output_file_path, "w") as output_file:
    for i, summary in enumerate(summaries):
        output_file.write(f"Summary {i+1}: {summary}\n")

print("Summary saved to:", output_file_path)
```

## GPT2Tokenizer

The GPT2Tokenizer question-answer pipeline first tokenizes the input question and context using the GPT-2 tokenizer, then feeds the tokenized sequence into a GPT-2 model that generates the answer based on the provided context and question.

```python
from transformers import pipeline, GPT2Tokenizer

# Load the question answering pipeline
qa = pipeline("question-answering")

# Load the text from the file
def load_text(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

text_path = '/content/drive/My Drive/dataset/sampleData/combined_output.txt'
text = load_text(text_path)

# Initialize the tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Set the maximum sequence length
max_length = 512

# Input questions
input_questions = [
    "What is the risk?",
    "What should be the length of the hose?",
    "What should we do if the softener has lost its fluidity?"
]

# Generate answers
for i, question in enumerate(input_questions):
    # Prepend the text with the question prompt
    prompt = question + " " + text

    # Truncate the prompt if it exceeds the maximum length
    encoded_prompt = tokenizer.encode(prompt, add_special_tokens=False)
    if len(encoded_prompt) > max_length:
        encoded_prompt = encoded_prompt[-max_length:]

    # Convert the encoded prompt back to text
    prompt = tokenizer.decode(encoded_prompt)

    # Answer the generated question
    answer = qa(question=question, context=prompt)

    # Print answers
    print(f"Question {i+1}: {question}")
    print(f"Answer: {answer['answer']}")
    print(f"Confidence: {answer['score']}")
    print()
```
